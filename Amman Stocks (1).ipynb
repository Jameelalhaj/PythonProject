# Data Pre Processing

```python
df2=df.iloc[2:]
display(df2.head())
```

```python
df2.columns=df2.iloc[0]
# Delete first row by selecting last n-1 rows
df3 = df2.tail(df2.shape[0] -1)
# Delete first column
df4 = df3.iloc[: , 1:]
# df3=df3.reset_index(drop=True)
display(df4.head())
```

```python
df4['Date'] = pd.to_datetime(df4['Date'], dayfirst=True)
display(df4.info())
```

```python
df5=df4.groupby(df4.Date.dt.month).tail(1) 
```

```python
display(df5.head())
```

```python
stocks=df5.set_index(df5.Date, inplace=False)
stocks=stocks.drop(columns=['Date'])
display(stocks.head())
```
